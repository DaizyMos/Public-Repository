```python
from collections import deque
class financeofficesystem:
    def __init__(self):
        self.queue = deque() #initialize an empty queue
    def add_student(self, student_name): #add a student to the queue(enqueue)
        self.queue.append(student_name)
        print(f"student {student_name} joined the line.")
    def serve_student(self): #serve (remove) the first student in the queue(dequeue)
        if not self.queue:
            print("no student in the queue to serve")
            return None
        served = self.queue.popleft()
        print(f"serving student: {served}")
        return served
    def display_queue(self): #display the current queue
        if not self.queue:
            print("the queue is currently empty")
        else:
            print("current queue:", list(self.queue))

office = financeofficesystem()
office.add_student("Alice")
office.add_student("Joe")
office.display_queue()
office.serve_student()
office.display_queue()

```
