```python
def linear_search(arr, target):
    for i in range(len(arr)):
               if arr[i]==target:
                       print("value {target} found at position {i}")
                       return i
                       print("value {target} not found at position {i}")
                       return -i
arr=[21, 10, 45, 32, 67, 5, 18]
target=32
linear_search (arr, target)
```

```python
def binary_search(arr, target):
    start=0
    end =len(arr)-1
    while start<=end:
          mid = (start + end) //2
          if arr[mid] == target: 
             print(f"Value {target} found at position {mid}")
             return mid
          elif arr[mid] < target:
               start = mid + 1
          else:  
              end = mid - 1
    print(f"Value {target} not found in the list")
    return -1

arr= [21, 10, 45, 32, 67, 5, 18]
arr.sort()
print(arr.sort)
target=45
binary_search(arr, target)
```

```python
def bubble_sort(arr):
    n=len(arr)
    for i in range(n):
        swapped=False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped=True
        if not swapped:
            break

arr=[21, 10, 45, 32, 67, 5, 18]
print("Original List:", arr)
bubble_sort(arr)
print("Sorted List (Ascending):", arr) 
```
